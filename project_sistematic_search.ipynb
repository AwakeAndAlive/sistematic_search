```python
#SISTEMATIC SEARCH - GOOGLE V0.1

import requests
from bs4 import BeautifulSoup

# URL 
query = "site:link.springer.com electoral redistricting genetic algorithm"
url = f"https://www.google.com/search?q={query}"

# Header necessário
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36"
}

# GET
response = requests.get(url, headers=headers)

if response.status_code == 200:
    soup = BeautifulSoup(response.text, 'html.parser')
    
    # Número de hits
    results_info = soup.find('div', {'id': 'result-stats'})
    results_text = results_info.text if results_info else 'No results info available'
    print(results_text)
    print()

    # Iterando para saída
    results = soup.find_all('div', class_='tF2Cxc')

    for result in results:
        # Título
        title_element = result.find('h3')
        title = title_element.text if title_element else 'No title available'
        
        # Link
        link_element = result.find('a', href=True)
        link = link_element['href'] if link_element else 'No link available'
        
        # Adicionais
        snippet_element = result.find('span', class_='aCOpRe')
        if snippet_element:
            additional_info = snippet_element.text
        else:
            # Coleta as demais
            additional_info_elements = result.find_all('span')
            additional_info = " ".join([element.text for element in additional_info_elements if element.text and 'class' not in element.attrs])
        
        # Limpando a saída
        additional_info = additional_info.replace(title, '').replace(link, '').strip()
        additional_info = additional_info.replace('https://link.springer.com', '').replace('Traduzir esta página', '').strip()
        
        # Impressão 
        print(f"Title: {title}")
        print(f"Link: {link}")
        print(f"Info: {additional_info}")
        print()
else:
    print("Failed to retrieve the webpage.")
```

```python
# SISTEMATIC SEARCH - GOOGLE SCHOLAR V0.1

import requests
from bs4 import BeautifulSoup

# URL 
query = "source:Springer electoral redistricting genetic algorithm"
url = f"https://scholar.google.com/scholar?hl=en&as_sdt=0%2C5&q={query}&btnG="

# Header necessário
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36"
}

# GET
response = requests.get(url, headers=headers)

if response.status_code == 200:
    soup = BeautifulSoup(response.text, 'html.parser')
    
    # Número de hits
    results_info = soup.find('div', {'id': 'gs_ab_md'})
    results_text = results_info.text if results_info else 'No results info available'
    print("Results found: " + results_text)
    print()

    # Iterando para saída
    results = soup.find_all('div', class_='gs_r gs_or gs_scl')

    for result in results:
        # Título
        title_element = result.find('h3', class_='gs_rt')
        title = title_element.text if title_element else 'No title available'
        
        # Link
        link_element = title_element.find('a', href=True) if title_element else None
        link = link_element['href'] if link_element else 'No link available'
        
        # Adicionais
        snippet_element = result.find('div', class_='gs_rs')
        additional_info = snippet_element.text if snippet_element else 'No additional info available'
        
        # Meta-informações (autores, ano, publicação)
        meta_element = result.find('div', class_='gs_a')
        meta_info = meta_element.text if meta_element else 'No meta info available'
        
        # Limpando a saída
        additional_info = additional_info.replace(title, '').replace(link, '').strip()
        
        # Impressão 
        print(f"Title: {title}")
        print(f"Link: {link}")
        print(f"Meta Info: {meta_info}")
        print(f"Info: {additional_info}")
        print()
else:
    print("Failed to retrieve the webpage.")
```
